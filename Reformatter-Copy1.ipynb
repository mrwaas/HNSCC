```python
FDA = set()
with open('io_files/protein_class_FDA.tsv', 'r') as fo:
    
    header = True
    
    for line in fo:
        
        if header:
            header = False
        
        else:
        
            line = line.rstrip()
        
            ensembl, gene, desc, ID, evidence = line.split('\t')
        
            FDA.add(ID)
            

            
potential = set()
with open('io_files/protein_class_Potential.tsv', 'r') as fo:
    
    header = True
    
    for line in fo:
        
        if header:
            header = False
        
        else:
        
            line = line.rstrip()

            ensembl, gene, desc, ID, evidence = line.split('\t')

            potential.add(ID)
            

            
drug_targets = dict()

for id in FDA:
    drug_targets[id] = "Approved"
    
for id in potential:
    drug_targets[id] = "Potential"
    

    
CD = dict()
with open('io_files/Human_CD_SP.csv', 'r') as fo:
    
    header = True
    
    for line in fo:
        
        if header:
            header = False
        
        else:
        
            line = line.rstrip()
        
            CD[line.split(',')[0]] = line.split(',')[1].rstrip()
            

SPC = dict()
with open('io_files/SPC_by_Source.csv', 'r') as fo:
    
    header = True
    
    for line in fo:
        
        if header:
            header = False
        
        else:
        
            line = line.rstrip()
        
            SPC[line.split(',')[0]] = int(line.split(',')[1])
            
            
gene = dict()
with open('io_files/gene_name.tsv', 'r') as fo:
    
    for line in fo:
        
        line = line.rstrip('\n')
        
        ID, x, gene_name = line.split('\t')
        
        if gene.get(ID):
            
            gene[ID].add(gene_name)
            
        else:
            
            gene[ID] = { gene_name }
```

```python
## this is to open a proteinGroup output from MQ and annotate it with additional information

with open('PG_step1.tsv', 'w') as fo2:

    with open('io_files/proteinGroups.txt', 'r') as fo:

        header = True

        for line in fo:
            
            
            if header:
                header = False
                line = line.rstrip('\n')
                line += '\t' + 'SPC' + '\t' + 'CD' + '\t' + 'DT' + '\n'
                fo2.write(line)

            else:
                line = line.rstrip('\n')
                sIDs = line.split('\t')[1]

                if sIDs.count(';') == 0:

                    if sIDs.startswith("REV"):
                        IDs = { }
                    elif sIDs.startswith("sp"):
                        IDs = { sIDs.split('|')[1].split('-')[0] }
                    elif sIDs.startswith("CON"):
                        IDs = { sIDs.split('_')[-1] }

                else:
                    ID_list = sIDs.split(';')
                    IDs = set()

                    for ID in ID_list:

                        if ID.startswith("sp"):
                            IDs.add( ID.split('|')[1].split('-')[0] )
                        elif ID.startswith("CON"):
                            IDs.add( ID.split('_')[-1] )

                dt_set = set()
                SPC_set = set()
                CD_set = set()

                for ID in IDs:

                    if ID in drug_targets:
                        dt_set.add(drug_targets[ID])

                    if ID in SPC:
                        SPC_set.add(SPC[ID])

                    if ID in CD:
                        CD_set.add(CD[ID])

                if dt_set:
                    dt = dt_set.pop()
                else:
                    dt = 'NA'

                if SPC_set:
                    spc = max(SPC_set)
                else:
                    spc = 0

                if CD_set:
                    cd = CD_set.pop()
                else:
                    cd = 'NA'
                

                line = line + '\t' + str(spc) + '\t' + cd + '\t' + dt + '\n'
                
                fo2.write(line)
```

```python
with open('PG_step2.tsv', 'w') as fo2:

    with open('PG_step1.tsv', 'r') as fo:

        header = True

        for line in fo:
                        
            if header:
                header = False
                line = line.rstrip('\n')
                line += '\t' + 'Concensus_Accession' +'\t' + 'Gene_Name' + '\n'
                
                fo2.write(line)
                

            else:
                line = line.rstrip('\n')
                sIDs = line.split('\t')[0]

                if sIDs.count(';') == 0:

                    if sIDs.startswith("REV"):
                        IDs = { }
                    elif sIDs.startswith("sp"):
                        IDs = { sIDs.split('|')[1].split('-')[0] }
                    elif sIDs.startswith("CON"):
                        IDs = { sIDs.split('_')[-1].split('-')[0] }

                else:
                    ID_list = sIDs.split(';')
                    IDs = set()
                    

                    for ID in ID_list:

                        if ID.startswith("sp"):
                            IDs.add( ID.split('|')[1].split('-')[0] )
                        elif ID.startswith("CON"):
                            IDs.add( ID.split('_')[-1].split('-')[0] )
                
                gene_names = set()
                
                if IDs:
                    for ID in IDs:
                        
                        if gene.get(ID):
                        
                            gene_names = gene_names.union(gene[ID])
                            
                        
                line += '\t' + ';'.join(IDs) + '\t' + ';'.join(gene_names)  + '\n'
                fo2.write(line)
                        
                
                
```
