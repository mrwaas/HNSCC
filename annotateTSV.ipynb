```python
gene = dict()


with open('io_files/gene_name.tsv', 'r') as fo: ####### change to correct locations
    
    for line in fo:
        
        line = line.rstrip('\n')
        
        ID, x, gene_name = line.split('\t')
        
        if gene.get(ID):
            
            gene[ID].add(gene_name)
            
        else:
            
            gene[ID] = { gene_name }
```

```python
## all gene names associated with IDs are written

in_path = ''
file_in = ''

out_path = ''
file_out = ''

with open( out_path + file_out , 'w') as fo2:

    with open( in_path + file_in , 'r') as fo:

        header = True

        for line in fo:
                        
            if header:
                header = False
                line = line.rstrip('\n')
                line += '\t' + 'Concensus_Accession' +'\t' + 'Gene_Name' + '\n'
                
                fo2.write(line)
                

            else:
                line = line.rstrip('\n')
                sIDs = line.split('\t')[0]

                if sIDs.count(';') == 0:

                    if sIDs.startswith("REV"):
                        IDs = { }
                    elif sIDs.startswith("sp"):
                        IDs = { sIDs.split('|')[1].split('-')[0] }
                    elif sIDs.startswith("CON"):
                        IDs = { sIDs.split('_')[-1].split('-')[0] }

                else:
                    ID_list = sIDs.split(';')
                    IDs = set()
                    

                    for ID in ID_list:

                        if ID.startswith("sp"):
                            IDs.add( ID.split('|')[1].split('-')[0] )
                        elif ID.startswith("CON"):
                            IDs.add( ID.split('_')[-1].split('-')[0] )
                
                gene_names = set()
                
                if IDs:
                    for ID in IDs:
                        
                        if gene.get(ID):
                        
                            gene_names = gene_names.union(gene[ID])
                            
                        
                line += '\t' + ';'.join(IDs) + '\t' + ';'.join(gene_names)  + '\n'
                fo2.write(line)
                        
                
                
```
